Stemming is the process of shortening a word to its base or root form, removing any suffixes or prefixes. This helps in understanding and processing natural language by simplifying words to their core meaning. ### Snowball Stemmer
The Snowball Stemmer, also known as the Porter2 Stemmer, is a tool used to reduce words to their base or root form. Imagine you have the words "running," "runner," and "ran." The Snowball Stemmer would simplify all these words to "run."

It's an improved version of the original Porter Stemmer. The Snowball Stemmer fixes some problems with the Porter Stemmer, making it more accurate in figuring out the root of words. Think of it as an updated, better-performing version of a word-reducing tool!

# Lemmatizer

Lemmatization is a lot like stemming, but it's a bit smarter. While stemming chops off the ends of words to get to the root, lemmatization gets you the actual base form of a word, known as the "lemma." The lemma is a real, valid word. The dictionary used for lemmatization, called WordNet, contains a lot more words compared to the lists used in stemming. This is one reason why lemmatization often gives more accurate and meaningful results than stemming.

# BAG OF WORDS

Binary Bag of Words: A representation where each word in the vocabulary is marked as 1 if it appears in the document, regardless of the number of times it appears.

Bag of Words (BOW): A representation where each word in the vocabulary is counted for its occurrences in the document, reflecting the frequency of each word.

Advantages:
Simple and Easy to Use: Easy to implement and understand.

Fixed Size Vector: Creates fixed-size vectors regardless of document length, which simplifies computations.


Disadvantages:
Sparsity: Results in sparse matrices, which can lead to inefficiencies and increased storage requirements.

Overfitting: High dimensionality can cause models to overfit, especially with large datasets.

Order Ignored: The sequence of words is not captured, losing the context of word positions.

Out of Vocabulary (OOV): New or unseen words in test data are not represented in the vocabulary.

Semantic Meaning Lost: Fails to capture the meanings and relationships between words.

# TF-IDF term frequency and inverse document frequency

Advantages of TF-IDF:

Easy to Use and Intuitive: Simple to implement and understand, commonly used in text analysis.

Fixed Size Vector: Produces a fixed-size feature vector based on the vocabulary size, making it manageable for machine learning models.

Captures Word Importance: Reflects the importance of words in a document relative to the corpus, helping to highlight relevant terms.


Disadvantages of TF-IDF:

Sparsity: Results in sparse matrices, where most values are zero, which can lead to inefficiencies in computation and storage.

Out of Vocabulary (OOV): New or unseen words in test data are not represented if they are not in the training vocabulary, limiting the model's ability to handle new terms.

# Word Embeddings

Word embedding is a technique in Natural Language Processing (NLP) that represents words as dense, continuous vectors of fixed size, capturing semantic meaning and relationships between words. Unlike one-hot encoding, word embeddings encode words based on their context and usage, facilitating better performance in various NLP tasks. Word embedding is essentially the process of converting words into continuous, dense vectors. These vectors capture semantic relationships and meanings based on the context in which words appear, allowing similar words to have similar vector representations. This is used in various NLP tasks to improve performance and understanding of language.

# Word2vec

Word2Vec is a popular word embedding technique that uses neural networks to learn vector representations of words from large text corpora. It captures semantic relationships by placing similar words closer together in the vector space. Word2Vec has two models: Continuous Bag of Words (CBOW) and Skip-gram.

Word2Vec models are trained on large text corpora by predicting words based on their surrounding context. There are two main architectures for this:

Continuous Bag of Words (CBOW): Predicts the target word from the surrounding context words.
Skip-gram: Predicts the surrounding context words given the target word.


Word2Vec transforms words into continuous vectors that reflect their meanings based on context, enabling more effective processing of textual data.

Cosine similarity is a measure used to determine how similar two vectors are, based on the cosine of the angle between them. It is commonly used in text analysis and natural language processing to compare the similarity between documents or word vectors.

How Cosine Similarity Works:
Vector Representation: Represent each text or word as a vector in a multi-dimensional space. For instance, in Word2Vec, each word is a vector of numbers.

Dot Product: Compute the dot product of the two vectors. The dot product measures the extent to which two vectors point in the same direction.

Magnitude (Norm): Calculate the magnitude (or length) of each vector. The magnitude is given by the square root of the sum of the squared components of the vector.




Advantages of Word2Vec:

Sparse to Dense Matrix: Converts sparse matrices into dense vectors.

Semantic Information: Captures semantic meanings (e.g., "king" - "man" + "woman" ≈ "queen").

Fixed Dimensions: Transforms vocabulary into fixed-size vectors.

Handles OOV Words: Can address out-of-vocabulary (OOV) issues with similar contexts.

Improve the accuracy of CBOW or Skip-gram, you can consider the following strategies:

Increase Training Dataset: More data generally leads to better generalization.

Increase Window Size: A larger window captures more context.

Increase Vector Dimensions: Higher dimensional vectors capture more information.

Adjust Hyperparameters: Tune learning rate, epochs, and other parameters.

Use Negative Sampling or Hierarchical Softmax: These methods can speed up training without significantly reducing accuracy.

### AvgWord2vec

It is a simple method that represents a word by averaging the vectors of all the words in a sentence or document. For example, if a sentence is "I love ice cream," it takes the vectors for "I," "love," "ice," and "cream," averages them, and uses this average vector to represent the sentence. This approach captures the overall meaning without needing complex models.

GloVe is ideal for quickly using pre-trained embeddings.
